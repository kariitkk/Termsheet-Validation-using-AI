**Reading TEXT from Images using OCR**

```python
import fitz  # PyMuPDF for PDF processing
import pytesseract  # OCR for image-based text extraction
from PIL import Image  # Image processing
import io

# Set Tesseract path if needed (Windows users)
# pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

def extract_text_from_pdf(pdf_path):
    """Extracts text from a PDF file, including OCR for scanned pages."""
    text = ""
    with fitz.open(pdf_path) as doc:
        for page in doc:
            page_text = page.get_text("text")
            if page_text.strip():  # If text is found, add it
                text += page_text + "\n"
            else:  # If no text, use OCR on images
                image_list = page.get_images(full=True)
                for img_index, img in enumerate(image_list):
                    xref = img[0]
                    base_image = doc.extract_image(xref)
                    image_bytes = base_image["image"]
                    img = Image.open(io.BytesIO(image_bytes))
                    img = img.convert("L")  # Convert to grayscale for better OCR
                    text += pytesseract.image_to_string(img) + "\n"
    return text

# Example Usage
pdf_file ="Doc2sampleOCR.pdf"# Replace with your file
pdf_text = extract_text_from_pdf(pdf_file)

print("Extracted Text from PDF:")
print(pdf_text[:])  # Print first 500 characters
```

```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text()
    return text

pdf_path = "termsheet02.pdf"  # Replace with your file path
text = extract_text_from_pdf(pdf_path)
print(text[:])  # Preview first 1000 characters
```

```python
import spacy
import re

nlp = spacy.load("en_core_web_sm")
doc = nlp(text)

# --- 1. Extract Key Entities ---
dates = list(set([ent.text for ent in doc.ents if ent.label_ == "DATE"]))  # Remove duplicates
percentages = list(set(re.findall(r"\d+%", text)))  # Unique percentages
orgs = list(set([ent.text for ent in doc.ents if ent.label_ == "ORG"]))  # Unique organizations

# --- 2. Extract Guarantee Rates (Improved Regex) ---
guarantee_rates = re.findall(r"(?:up to|maximum of)\s*(\d+)%", text, re.IGNORECASE)  # Catches "up to 80%"

# --- 3. Extract Funding Categories (More Robust) ---
categories = re.findall(r"Category\s+[A-D]:.*?(?=\n\n|$)", text, re.DOTALL | re.IGNORECASE)
if not categories:  # Fallback if regex fails
    categories = ["No categories found. Check PDF formatting."]

# --- 4. Print Results with Error Handling ---
print("=== Key Entities ===")
print("Dates:", dates[:5])  # First 5 unique dates
print("Percentages:", percentages)
print("Organizations:", orgs[:5])  # First 5 unique orgs

print("\n=== Guarantee Rates ===")
print(guarantee_rates if guarantee_rates else "No guarantee rates found.")

print("\n=== Funding Categories ===")
print(categories[0][:200] + "..." if categories else "No categories found.")
```

**SUMMARY OF THE TEXT** 

```python
from collections import Counter

def simple_summarize(text, num_sentences=3):
    sentences = [s.strip() for s in text.split('.') if s]
    word_freq = Counter(word.lower() for word in text.split() if word.isalpha())
    ranked_sentences = sorted(
        sentences,
        key=lambda s: sum(word_freq[word.lower()] for word in s.split() if word.isalpha()),
        reverse=True
    )
    return '. '.join(ranked_sentences[:num_sentences]) + '.'

print("Simple Summary:", simple_summarize(text))
```
